```python
datetimes = []
date_start =  datetime(2023, 6, 23, 11, 0, 1)
date_end =  datetime(2023, 6, 23, 18, 0, 0)
tz = timezone(timedelta(days=-1, seconds=72000), 'EDT')
n = 10000

for _ in range(n):

    dt = fake.future_datetime(
        end_date="+2h",
        tzinfo=tz
        )
    datetimes.append(dt)
datetimes
```

```python
df = (
    pd.DataFrame(datetimes, columns=['datetime'])
    .assign(
        color= lambda df_: [fake.color_name() for i in range(len(df_))]
    )
    .sort_values('datetime')
    .reset_index(drop=True)
)
df.head(100)
```
